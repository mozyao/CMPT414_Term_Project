```
import requests  
file_url = "http://www.cs.toronto.edu/~kriz/cifar-10-python.tar.gz"
    
r = requests.get(file_url, stream = True)  
root_folder = "/content/drive/My Drive/CMPT414_Project_setup"
with open(root_folder+"/dataset/cifar-10-python.tar.gz", "wb") as file:  
    for block in r.iter_content(chunk_size = 1024): 
         if block:  
             file.write(block)  
             
 dataset_path = root_folder+"/dataset/"
 %cd $dataset_path
 !tar xvzf cifar-10-python.tar.gz
```
